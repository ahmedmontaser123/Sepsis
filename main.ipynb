```python
import pandas as pd
import numpy as np
import plotly.express as px
import requests
```

```python
url = "https://raw.githubusercontent.com/ahmedmontaser123/Sepsis/refs/heads/master/train_merged_per_visit.csv"
```

```python
data = pd.read_csv(url)
```

```python
# check duplication and NaNs
print("Data shape:", data.shape)

# full-row duplicates
dup_all_count = data.duplicated().sum()
print("Duplicate rows (all columns):", dup_all_count)
if dup_all_count:
    print(data[data.duplicated(keep=False)].head().to_string())

# duplicates by visit_occurrence_id (common identifier)
if 'visit_occurrence_id' in data.columns:
    dup_id_count = data['visit_occurrence_id'].duplicated().sum()
    print("Duplicate visit_occurrence_id count:", dup_id_count)
    if dup_id_count:
        print(data[data['visit_occurrence_id'].duplicated(keep=False)].sort_values('visit_occurrence_id').head().to_string())

# missing values per column (count and percent)
missing_counts = data.isna().sum().sort_values(ascending=False)
missing_pct = (missing_counts / len(data) * 100).round(2)
missing_summary = pd.concat([missing_counts, missing_pct], axis=1)
missing_summary.columns = ['missing_count', 'missing_pct']
print("\nTop 20 columns by missing values:")
print(missing_summary.head(20).to_string())

# overall row-wise missing stats
row_missing = data.isna().sum(axis=1)
rows_with_any_nan = (row_missing > 0).sum()
print(f"\nRows with any NaN: {rows_with_any_nan} ({rows_with_any_nan/len(data)*100:.2f}%)")
print("Row-wise missing count distribution:")
print(row_missing.describe().to_string())

# quick lists
print("\nNumber of columns with no missing values:", (missing_counts==0).sum())
print("Columns with 100% missing values:", missing_counts[missing_counts==len(data)].index.tolist())
```

```python
df = pd.DataFrame(data)
```

```python
df['visit_start_date'] = pd.to_datetime(df['visit_start_date'])
df['birth_datetime'] = pd.to_datetime(df['birth_datetime'])
df = df.sort_values(by = 'visit_start_date')
```

```python
df['age_in_year'] = df['visit_start_date'].dt.year - df['birth_datetime'].dt.year
```

```python
df.head(10)
```

```python
# you have split data based on past (train) and future (prediction) l
cutof_date = df['visit_start_date'].quantile(0.8)
train = df[df['visit_start_date'] < cutof_date]
test = df[df['visit_start_date'] >= cutof_date]
train_patient_ids = set(train['person_id'])
test = test[~ test['person_id'].isin(train_patient_ids)]
print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
print(f"train data : {train['person_id'].duplicated().sum()}")
print(f"test data : {test['person_id'].duplicated().sum()}")
```

```python
# sum of all nulls
print(f'sum nulls in train: {train.isna().sum().sum()}')
print(f'sum nulls in test: {test.isna().sum().sum()}')
```

```python
cols = df.iloc[:, 7:].columns

train = train.copy()
test  = test.copy()

train = train.sort_values(['person_id', 'visit_start_date'])
test  = test.sort_values(['person_id', 'visit_start_date'])

train.loc[:, cols] = train.groupby('person_id')[cols].ffill()
test.loc[:, cols]  = test.groupby('person_id')[cols].ffill()
```

```python
# sum of all nulls
print(f'sum nulls in train: {train.isna().sum().sum()}')
print(f'sum nulls in test: {test.isna().sum().sum()}')
```

```python
train['SepsisLabel'].value_counts(normalize = True)
```

```python
test['SepsisLabel'].value_counts(normalize = True)
```

```python
x_train = train.drop(columns = ['visit_occurrence_id','person_id','visit_start_date','birth_datetime','age_in_months','SepsisLabel'])
y_train = train['SepsisLabel']
x_test = test.drop(columns = ['visit_occurrence_id','person_id','visit_start_date','birth_datetime','age_in_months','SepsisLabel'])
y_test = test['SepsisLabel']
```

```python
from clean import Cleaner
from processing import preprocessorCleaner
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report,recall_score


pipeline = Pipeline(steps = [
    ('cleaner', preprocessorCleaner()),
    ('SMOTE',  SMOTETomek(random_state = 42,sampling_strategy='minority')),
    ('model', XGBClassifier(scale_pos_weight = len(y_train[y_train== 0])+20 / len(y_train[y_train==1]),n_estimators = 300,max_depth = 3,gamma = 0.8,learning_rate = 0.01 ) )
])

```

```python
pipeline.fit(x_train,y_train)   
```

```python
y_pred = pipeline.predict(x_test)
print(recall_score(y_test , y_pred))
```

```python
print(classification_report(y_test , y_pred))
```
